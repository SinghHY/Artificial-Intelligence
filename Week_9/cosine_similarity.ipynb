```python
import numpy as np
from fractions import Fraction

user_A_input = input("Please enter A as a list (e.g., [2/3, 0,0,5/3,-7/3,0,0]): ")

# Remove brackets and split values
item_A = user_A_input.strip("[]").split(",")

# Convert each value
A_list = [float(Fraction(item)) for item in item_A]

A = np.array(A_list)

print("Input A as NumPy array:", A)
```
