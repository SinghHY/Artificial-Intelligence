```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten,MaxPooling2D
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train[0]
```

```python
# 2) Normalize pixels to [0,1]
X_train = X_train.astype("float32") / 255.0
X_test  = X_test.astype("float32") / 255.0
```

```python
# 3) Flatten 28x28 images into 784-length vectors (ANN needs 1D input)
X_train = X_train.reshape(-1, 28*28)
X_test  = X_test.reshape(-1, 28*28)
```

```python
# 4) Build ANN
model = keras.Sequential()
model.add(keras.layers.Input(shape=(784,)))
model.add(keras.layers.Dense(128, activation="relu"))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dense(10, activation="softmax"))  # 10 classes (0-9)
```

```python
model.summary()
```

```python
# ------------------------------------------------
# Compile model
# ------------------------------------------------
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# ------------------------------------------------
# Train model
# ------------------------------------------------
history = model.fit(
    X_train, y_train,
    epochs=1,
    batch_size=128,
    validation_split=0.1
)
```

```python
# ------------------------------------------------
# Training accuracy (last epoch)
# ------------------------------------------------
train_accuracy = history.history['accuracy'][-1]
print("\nFinal Training Accuracy:", train_accuracy)
```

```python
# ------------------------------------------------
# Test model
# ------------------------------------------------
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)

print("Final Test Accuracy:", test_accuracy)
```
